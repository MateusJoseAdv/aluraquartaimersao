# ***Criando Estruturas de Dados***

```python
# Vou criar uma lista simples que contém uma sequência numérica de 1 à 5.
```

```python
import pandas as pd
data = [1,2,3,4,5]
```

```python
# Irei nomear essa Series de "random", chamar o Pandas (pd) e passaremos "data":

random = pd.Series(data)
```

```python
# Irei Escrever na próxima célula "random" para visualizar a série.
# Terei na primeira coluna um índice de 0 à 4 e nossa sequência.

random
```

```python
# Tenho algumas outras opções para passar para Series, o index é uma delas. 
# Escreverei exatamente este termo: "index". Em seguida farei uma "list comparation" e escreverei "Linha". 
# Adicionarei ainda, "+ str(i) for i in range(5)".

index = ['Linha' + str(i) for i in range(5)]
index
```

```python
# Esses conteúdos foram armazenados na variável "index", e passarei essa variável como índice da minha Series:

random = pd.Series(data = data, index = index)
random 
```

```python
# Vou agora efetivamente criar a Series:

random = pd.Series(data)
s1 = random + 2
s1
```

```python
# Desse modo, foi somado o valor 2 em cada uma das linhas.
# Posso somar, dividir ou multiplicar duas Series que possuam os mesmos índices.
# Portanto, criarei um "s2", que será composto por "random" + "s1"

s2 =  random + s1
s2
```

```python
# Isto posto, caso seja somada uma linha que não seja compatível, será obtido um valor nulo.
```

```python
# Irei trabalhar com uma nova lista numérica - uma lista que contêm listas.

data = [[1, 2, 3], 
        [4, 5, 6], 
        [7, 8, 9]]
```

```python
# Passarei esta lista para a função DataFrame, usando a variável "df1" :

df1 = pd.DataFrame(data = data)
df1
```

```python
# Preciso agora criar os rótulos para cada linha utilizando o index. 
# Vou criar uma lista chamada de "Linha", e então faremos o mesmo procedimento de concatenação que realizado na lista anterior.

index = ['Linha' + str(i) for i in range(3)]
index
```

```python
# Para inserir essa lista em nosso DataFrame, escreverei:

df1 = pd.DataFrame(data = data, index = index)
df1
```

```python
# Farei o mesmo procedimento para as colunas (columns). Utilizarei uma estrutura de código parecida com a da lista "Linha",
# mas dessa vez trocaremos esse termo por "Colunas".

columns = ['Coluna' + str(i) for i in range(3)]
columns
```

```python
# Partiu inserir estes rótulos em nosso DataFrame, que chamaremos de "df2":

df2 = pd.DataFrame(data = data, index = index, columns = columns)
df2
```

```python
# Está criado um DataFrame em formato padrão. Posso obter o mesmo resultado utilizando a metodologia de dicionário que
# utilizamos na Series, então se escrever:

data = {'Coluna0': {'Linha0': 1, 'Linha1': 4, 'Linha2': 7},
        'Coluna1': {'Linha0': 2, 'Linha1': 5, 'Linha2': 8},
        'Coluna2': {'Linha0': 3, 'Linha1': 6, 'Linha2': 9}}
```

```python
# O resultado será o mesmo DataFrame que gerei, perceba.
# Posso, ainda, atingir o mesmo resultado por meio de uma lista de tuplas ao inserir parênteses:

data = [(1, 2, 3), 
        (4, 5, 6), 
        (7, 8, 9)]
data
```

```python
# Que tal criar um novo DataFrame com base nesta lista? Irei a nomear de "d3".
# Irei inserir o "index" e as "columns" para obter um resultado mais interessante:

df3 = pd.DataFrame(data = data, index = index, columns = columns)
df3
```

```python
# Em momento posterior, concatenarei DataFrames. Mas, por ora, farei uma modificação simples em "df1". 
# Para os valores maiores que "0", será atribuída a nomeação "A".

df1[df1 > 0] = 'A'
df1
```

```python
# Farei o mesmo procedimento para o DataFrame "df2" e "df3", mas agora utilizarei a letra "B" e "C", respectivamente

df2[df2 > 0] = 'B'
df2
```

```python
df3[df3 > 0] = 'C'
df3
```

```python
# Feito isso, irei acionar a ferramenta "concat()"" do Pandas, que como indica o nome, realiza concatenações. 
# Passarei a lista dos DataFrames que deverão ser concatenados. Meu objetivo inicial é empilhar as tabelas, 
# de modo que a junção entre elas sejam as colunas.

# Gerarei, então, um "d4", que me exibirá a concatenação de DataFrames:

df4 = pd.concat([df1, df2, df3])
df4
```

```python
# É interessante notar que posso ter rótulos iguais para linhas diferentes. 
# Também posso fazer a ligação entre tabelas se dar pelas linhas, e não pelas colunas. Para isso, escreverei:
```

```python
df4 = pd.concat([df1, df2, df3], axis = 1)
df4
```

```python
# Irei ilustrar outros usos do concat() durante o curso, mas vi de forma rápida o que ocorre com um 
# DataFrame ao realizarmos uma concatenação.
```
