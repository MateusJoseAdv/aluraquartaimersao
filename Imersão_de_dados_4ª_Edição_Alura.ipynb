```python
len(dados["Valor"].str.split(expand = True)[1].unique())
```

*Existem, portanto, 10.008 linhas com somente 1038 preços, o que significa dizer que há muitos preços iguais/repetidos.*

```python
dados["Valor"].str.split(expand = True)[2].unique()
```

*Repetida a função de análise da coluna 0 para a coluna 2, percebe-se que nesta existem as informações de mês, ano e dia. Isto quer dizer que, provalvelmente, há dados de venda e dados de aluguel.*
*Precisamos agora migrar os dados do split para o dataframe:*

```python
dados [["Moeda", "Valor_anuncio", "Tipo_anuncio"]] = dados["Valor"].str.split(expand = True)
dados.head()
```

*É possível selecionar uma única coluna (ex. Valor), bem como selecionar mais de uma coluna. Será necessário passar uma lista com o nome de cada coluna que se deseja manipular.*

*Deste modo, para continuar o tratamento de dados, será necessário eliminar todo tipo de anúncio que não está vazio:*

```python
dados[dados["Tipo_anuncio"].isnull()]["Tipo_anuncio"].unique()
```

*Assim, se percebe a existência de um array vazio - o que demonstra que o pretendido deu certo.*

```python
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]
dados_vendas
```

*Após a análise, restou evidente que há somente 9.159 imóveis disponíveis para venda. Estão excluídas as locações e afins.*

# **Desafios 3 e 4 - Informar o preço do metro quadrado por bairro na cidade de São Paulo, demonstrando-os em gráficos.**

```python
valores_bairro = dados_vendas.groupby('Bairro')['Valor_anuncio_float'].sum()
metragem_bairro = dados_vendas.groupby('Bairro')['Metragem'].sum()

valor_m2_2 = valores_bairro/metragem_bairro
valor_m2_2 = valor_m2_2.sort_values(ascending=False)
plt.figure(figsize=(15, 8))
valor_m2_2[:30].plot(kind='bar', ec = "black", color = "grey")

plt.title('Valor do m² dos trinta bairros mais caros de São Paulo', pad = 22, fontsize = 22)
plt.ylabel('Valor do m² em reais')
plt.show()
```

*A fim de facilitar o entendimento, visto que existem muitos bairros no nosso dataset, fora calculado o preço do metro quadrado dos trinta bairros mais caros de São Paulo. Após a conclusão da análise de dados, percebe-se que o imovel existente no bairro Vila Guaca é um exemplo de outlier, uma vez que o valor de seu m² é discrepante face o restante dos bairros.
Após a filtragem minuciosa deste outlier, nota-se a existência de somente um imóvel nesta localidade, apresentando os seguintes dados:*

```python
dados_vendas[dados_vendas['Bairro'] == 'Vila Guaca']
```

*Este imóvel apresenta a pífia metragem de apenas 10m², supostamente contendo 3 quartos, 2 banheiros e 2 vagas de garagem. Além disso, está sendo anunciado pelo valor total de oitocentos mil reais, gerando o valor absurdo de oitenta mil reais por m². Isto posto, resta evidente a incongruência de dados, fato que gera a distorção apresentada no gráfico.*
