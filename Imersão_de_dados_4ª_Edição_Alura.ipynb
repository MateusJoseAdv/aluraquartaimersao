## Desafio 4 - Informar outros dados presentes no database, tais como a mediana, a média, a mínima e a máxima.

```python
dados_numericos = dados[["Metragem", "Quartos", "Banheiros", "Vagas"]]

dados_media = round(dados_numericos.mean(), 1)
dados_mediana = dados_numericos.median()
dados_min = dados_numericos.min()
dados_max = dados_numericos.max()
```

```python
calculos = {"Média" : dados_media, "Mediana" : dados_mediana, "Mínimo" : dados_min, "Máximo" : dados_max}
estatisticas = pd.DataFrame(calculos)
estatisticas
```

```python
dados["Valor"].str.split(expand = True)
```

*Deste modo foram criados dataframes com cada elemento da separação (apenas três elementos separados por dataframes).*

*O tratamento de dados acabou exigindo a necessidade de descobrir se todos os elementos da coluna 0 contêm, de fato, somente (R$)*:

```python
dados["Valor"].str.split(expand = True)[0].unique()
```

*Por sorte não havia outro dado na coluna 0 senão o cifrão desejado (R$), o que a tornou limpa.*

*É possível também descobrir quantos valores (preços) existem na coluna 1:*
*Obs. len marca a extensão da array.*

```python
len(dados["Valor"].str.split(expand = True)[1].unique())
```

*Existem, portanto, 10.008 linhas com somente 1038 preços, o que significa dizer que há muitos preços iguais/repetidos.*

```python
dados["Valor"].str.split(expand = True)[2].unique()
```

*Repetida a função de análise da coluna 0 para a coluna 2, percebe-se que nesta existem as informações de mês, ano e dia. Isto quer dizer que, provalvelmente, há dados de venda e dados de aluguel.*
*Precisamos agora migrar os dados do split para o dataframe:*

```python
dados [["Moeda", "Valor_anuncio", "Tipo_anuncio"]] = dados["Valor"].str.split(expand = True)
dados.head()
```

*É possível selecionar uma única coluna (ex. Valor), bem como selecionar mais de uma coluna. Será necessário passar uma lista com o nome de cada coluna que se deseja manipular.*

*Deste modo, para continuar o tratamento de dados, será necessário eliminar todo tipo de anúncio que não está vazio:*

```python
dados[dados["Tipo_anuncio"].isnull()]["Tipo_anuncio"].unique()
```

*Assim, se percebe a existência de um array vazio - o que demonstra que o pretendido deu certo.*

```python
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]
dados_vendas
```

*Após a análise, restou evidente que há somente 9.159 imóveis disponíveis para venda. Estão excluídas as locações e afins.*

*Agora é preciso transformar o "valor_anuncio" em números.*

```python
dados_vendas.info()
```

```python
dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas["Valor_anuncio_float"] = dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas.info()
```

```python
dados_vendas["Valor_anuncio_float"].plot.hist(bins = 50)
```
